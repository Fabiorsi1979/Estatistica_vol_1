# Intervalo de confiança para uma média

## Cálculo que expressa a variação em uma amostra (variação Inferior/ Superior)
#### Abaixo dois exemplos aplicados a uma amostra e a uma população, sendo exemplos de alunos de uma escola com a média de 1,65m e desvio padrão de 0,10m e 95% de intervalo de confiança

IC-t.jpg

```python
from math import sqrt
from scipy.stats import t

x= 1.65
s = 0.10
n = 30
grau_confianca = 95
t = t.ppf(1 - ((1-(grau_confianca/100))/2), df=(n-1))
Erro_confianca = s/(sqrt(n))

print(f'IC = {x - (t * Erro_confianca):.2f}, {x + (t * Erro_confianca):.2f}' )
```

# Teste de Proporção



### Uma maneira de verificar se uma proporção em uma amostra é diferente (ou igual) a uma proporção conhecida em toda população.

##### Suponha que numa escola a proporção esperada é de que 50% dos alunos gostam de esportes, foi feito uma pesquisa amostral de 100 alunos e você quer saber se mais ou menos 50% destes alunos gostam de esportes.

Teste_proporcao.jpg

```python
from math import sqrt

total_entrevistado = 100
resposta_positiva = 60
P_observada = resposta_positiva/total_entrevistado
P_esperada = 0.5
n = total_entrevistado

print(f' z = {(P_observada - P_esperada) / sqrt((P_esperada * (1 - P_esperada)) / n):.2f}')
```
